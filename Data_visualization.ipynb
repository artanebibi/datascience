```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/stoltzmaniac/wine-reviews-kaggle/refs/heads/master/winemag-data_first150k.csv")
```

```python
df # no preprocessing on this data set in this file. This is only for testing my visualization skills in python
```

```python
sns.histplot(df['country']) # df['country'].plot.hist() with pandas but you need numerical data to use hist() Seaborn is faster and better
```

```python
sns.histplot(df['points'], bins = 15)
sns.histplot(df['points'], binrange = (85, 95), color='red', bins = 10)
```

```python
sns.displot(df['price'], kde = True, height=5)
```

```python
country_price_mean = df.groupby('country')['price'].mean()
```

```python
country_price_mean
```

```python
country_price_mean.plot.bar()
```

```python
plt.plot(df['price'], df['points'].sort_values())
```

```python
# heatmap needs all of the feature values to be numbers so need to encode all of them. NOT TAKING NOTE WHAT
# ENCODING IS DONE. THIS IS USED ONLY FOR VISUALIZATION
encoder = LabelEncoder()
for feature in df.columns:
  df[feature] = encoder.fit_transform(df[feature])
```

```python
df
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True, cmap='icefire')
plt.show()
```
